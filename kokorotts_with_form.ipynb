```python
!pip install -q kokoro>=0.9.4 soundfile
!apt-get -qq -y install espeak-ng > /dev/null 2>&1
from kokoro import KPipeline
from IPython.display import display, Audio
import soundfile as sf
import torch
pipeline = KPipeline(lang_code='a')
text = '''

What Your Hair Says About Your Health
'''
generator = pipeline(text, voice='af_sky')
for i, (gs, ps, audio) in enumerate(generator):
    print(i, gs, ps)
    display(Audio(data=audio, rate=24000, autoplay=i==0))
    sf.write(f'{i}.wav', audio, 24000)
```

```python
from IPython.display import HTML, display
from google.colab import files
import base64

text = "Glory be to God in the most high" # @param {"type":"string"}
filename = "1.wav" # @param {"type":"string"}
voice = "af_sky" # @param {"type":"string"}
generator = pipeline(text, voice=voice)
for i, (gs, ps, audio) in enumerate(generator):
    print(i, gs, ps)
    display(Audio(data=audio, rate=24000, autoplay=i==0))
    sf.write(f'{filename}', audio, 24000)

# Generate a downloadable link
def create_download_link(filename, text):
    b64 = base64.b64encode(open(filename, 'rb').read()).decode()
    href = f'<a href="data:file/wav;base64,{b64}" download="{filename}">{text}</a>'
    return href

# Display the download button
display(HTML(create_download_link(filename, f'Download {filename}')))
```
